```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def plot_final(x, y_f, y_g, label):
    plt.figure(figsize=(4, 3))
    plt.plot(x, y_f, label=label)
    plt.plot(x, y_g, '--', label=f"Imamototang approx: {label}")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
start_time = time.time()
g, epsilons = imamototang.immamoto_optimal_approximation(np.exp, np.exp, 0, 1, 10)
print("End time e(x) [in seconds] ", round((time.time()-start_time), 3))
print("Mean epsilons: ", round(np.mean(epsilons), 5))

x = np.linspace(0, 1, 1000, endpoint=True)
y_f = np.array([np.exp(x_) for x_ in x])
y_g = np.array([g(x_) for x_ in x])

plot_final(x, y_f, y_g, label=r"$e^{x}$")
```

```python
start_time = time.time()
g, epsilons = imamototang.immamoto_optimal_approximation(
    lambda x: np.power(x, 2), lambda x: 2 * x, -1, 1, 10
)

print("End time x**2 [in seconds] ", round((time.time() - start_time), 3))
print("Mean epsilons: ", round(np.mean(epsilons), 5))

x = np.linspace(-1, 1, 1000, endpoint=True)
y_f = np.array([np.power(x_, 2) for x_ in x])
y_g = np.array([g(x_) for x_ in x])

plot_final(x, y_f, y_g, label=r"$x^2$")
```

```python
start_time = time.time()
g, epsilons = imamototang.immamoto_optimal_approximation(
    lambda x: np.power(x, 3), lambda x: 3 * (x**2), 0, 1, 10
)
print("End time x**3 [in seconds] ", round((time.time() - start_time), 3))
print("Mean epsilons: ", round(np.mean(epsilons), 5))

x = np.linspace(0, 1, 1000, endpoint=True)
y_f = np.array([np.power(x_, 3) for x_ in x])
y_g = np.array([g(x_) for x_ in x])

plot_final(x, y_f, y_g, label=r"$x^3$")
```
