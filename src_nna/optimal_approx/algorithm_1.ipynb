```python
import numpy as np
import matplotlib.pyplot as plt
import time
import sympy as sp
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
x_sym = sp.Symbol("x")
```

```python
def plot_final(x, y_f, y_g, label):
    plt.figure(figsize=(4, 3))
    plt.plot(x, y_f, label=label)
    plt.plot(x, y_g, '--', label=f"Optimal approx: {label}")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
start_time = time.time()


exp_fn = alg1.exp
g, optimal_approx_errors = alg1.optimal_approximation_algorithm(exp_fn, 0, 1, 10)
print("End time e(x) [in seconds] ", round((time.time()-start_time), 3))
print(u"Mean ΔS: ", round(np.mean(optimal_approx_errors), 5))

x = np.linspace(0, 1, 1000, endpoint=True)
y_f = np.array([np.exp(x_) for x_ in x])
y_g = np.array([g.subs(x_sym, x_) for x_ in x])

plot_final(x, y_f, y_g, label=r"$e^{x}$")
```

```python
start_time = time.time()

x_sq = alg1.x_sq
g, optimal_approx_errors = alg1.optimal_approximation_algorithm(x_sq, -1, 1, 10)
print("End time x**2 [in seconds] ", round((time.time() - start_time), 3))
print("Mean ΔS: ", round(np.mean(optimal_approx_errors), 5))

x = np.linspace(-1, 1, 1000, endpoint=True)
y_f = np.array([x_sq(x_) for x_ in x])
y_g = np.array([g.subs(x_sym, x_) for x_ in x])

plot_final(x, y_f, y_g, label=r"$x^2$")
```

```python
start_time = time.time()

x_cube = alg1.x_cube
g, optimal_approx_errors = alg1.optimal_approximation_algorithm(x_cube, 0, 1, 10)
print("End time x**3 [in seconds] ", round((time.time() - start_time), 3))
print("Mean ΔS: ", round(np.mean(optimal_approx_errors), 5))

x = np.linspace(0, 1, 1000, endpoint=True)
y_f = np.array([x_cube(x_) for x_ in x])
y_g = np.array([g.subs(x_sym, x_) for x_ in x])

plot_final(x, y_f, y_g, label=r"$x^3$")
```
