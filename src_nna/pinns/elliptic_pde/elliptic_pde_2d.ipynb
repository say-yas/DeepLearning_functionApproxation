# Summary of results for solving elliptic differential equation by PINNS 

The equation we solve read $ \frac{\partial^{2}}{ \partial^{2} x} u(x,y)
    +\frac{\partial^{2}}{ \partial^{2} y} u(x,y) =0.$

```python
import time
import torch

from scipy.stats import qmc
import numpy as np

from  src_nna.pinns.nn_pinns import MLP_PINN
from src_nna.pinns.elliptic_pde import elliptic_pde_2d
from mpl_toolkits.axes_grid1 import make_axes_locatable

import matplotlib.pyplot as plt
```

For the boundary condition being a constant function $c=3$, we experiment with the complexity of the model. 

```python
plotting = True
n_bc = 4
n_data_per_bc = 2000
Lx=1.0
Ly=1.0 

Nc = 1000 
n = 700

num_hidden=12
num_layers=8

lear_rate=  5e-3
num_iteration = 5000000 

function_bound = elliptic_pde_2d.cnst

for num_layers in [4, 6, 8, 10]:
    for j in range(4):
        num_hidden = np.power(2, 3+j)
        print("# hidden layers:", num_layers)
        print("#units in hidden layers:", num_hidden)

        elliptic_pde_2d.run_traning(n_bc, Lx, Ly, Nc, n, num_hidden, num_layers,
                        lear_rate, num_iteration, function_bound,
                          n_data_per_bc, plotting = False, verbose=False)
        
```
